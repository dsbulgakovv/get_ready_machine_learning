# Math

### Какие тесты используются для сравнения выборок?
Для сравнения выборок применяются различные статистические тесты в зависимости от типа данных, характера распределения и конкретной задачи анализа. Основные тесты включают:

#### Т-тест Стьюдента:
Используется для сравнения средних двух выборок (когда выборки маленькие, n<30). Применяется при условии нормальности распределения данных и приблизительно равной дисперсии в группах.
Есть несколько видов:
Парный т-тест: для зависимых выборок.
Независимый т-тест: для независимых выборок.
Одновыборочный т-тест: для сравнения среднего выборки с известным значением.
### U-тест Манна–Уитни:
Непараметрический тест для сравнения двух независимых выборок.
Используется, когда данные не соответствуют нормальному распределению.
### Z-тест:
Используется для проверки гипотез относительно средних или пропорций в больших выборках.
Предполагает, что выборки подчиняются нормальному распределению, и выборка большая (обычно n>30).
### Критерий χ² (хи-квадрат):
Применяется для категориальных данных.
Критерий согласия: проверяет, соответствуют ли распределения наблюдаемых и ожидаемых значений.
Критерий независимости: проверяет взаимосвязь между двумя категориальными переменными.
Какие методы проверки распределений на нормальность существуют?
### Тест Шапиро–Уилка (статистический метод):
Подходит для небольших и средних выборок (до 2000 наблюдений).
Гипотеза: если p-значение больше выбранного уровня значимости (например, 0.05), можно предположить, что распределение нормальное.
### Тест Колмогорова–Смирнова (статистический метод):
Применяется для проверки соответствия данных нормальному распределению.
Менее точен для небольших выборок и чувствителен к форме распределения.
Используется реже, чем тест Шапиро-Уилка, так как может быть менее мощным.
### Q-Q Plot (график квантиль-квантиль, графический метод):
Построение квантилей данных против квантилей нормального распределения.
Если данные нормальны, точки будут располагаться вдоль линии.
Один из наиболее наглядных методов, позволяющий быстро оценить отклонения от нормальности.

### Общее
- Уровень значимости (α): Обычно выбирается 5% (или 0,05), что означает, что вероятность ошибочно отклонить нулевую гипотезу (ложноположительный результат) равна 5%.
- Мощность теста (1−β): Мощность теста обычно составляет 80% или 90%. Она показывает вероятность обнаружения реального эффекта, если он существует.
- Минимальный детектируемый эффект (MDE): Это минимальный уровень различий, который необходимо обнаружить. Например, если мы хотим обнаружить улучшение конверсии с 5% до 5.5%, то MDE будет 0.5%. Чем меньше эффект, который мы хотим обнаружить, тем больше выборка.
- Базовая конверсия: Текущий уровень конверсии для варианта A. Знание этого помогает точно рассчитать нужный размер выборки.
- Что такое ошибка первого и второго рода?
Ошибка первого рода: нулевая гипотеза отвергается, хотя она на самом деле верна.
Ошибка второго рода: нулевая гипотеза принимается, хотя на самом деле она ложна.
- Что такое p-value и как его интерпретировать?
P-value — это вероятность получения таких же (или более экстремальных) результатов, как в наших данных, при условии, что нулевая гипотеза верна. В других словах, это мера того, насколько наблюдаемые результаты соответствуют нулевой гипотезе. Чем меньше p-value, тем менее вероятно, что наши данные согласуются с нулевой гипотезой.
Пример интерпретации:
Если p-value составляет 0.03, и мы выбрали уровень значимости 0.05, это означает, что существует только 3% вероятность получить такие данные (или более экстремальные), если нулевая гипотеза верна. Это позволяет нам отвергнуть нулевую гипотезу и заключить, что эффект вероятно существует.
Если p-value = 0.08 при уровне значимости 0.05, то у нас нет достаточных доказательств для отклонения нулевой гипотезы, так как вероятность получить такие результаты случайно составляет 8%.
- Как рассчитать мощность статистического критерия?
Мощность теста помогает избежать ошибки второго рода и обеспечивает уверенность в том, что тест достаточно чувствителен для выявления значимого эффекта.
Мощность выбирается исходя из уровня значимости, MDE, размера выборки и стандартного отклонения в данных.
Чем больше выборка, тем выше мощность теста.
Чем выше разброс значений, тем сложнее обнаружить значимый эффект, поэтому потребуется больший размер выборки для достижения той же мощности.

- Что такое центральная предельная теорема (ЦПТ) и закон больших чисел (ЗБЧ)?
- Центральная предельная теорема (ЦПТ): при достаточно большом размере выборки распределение средних значений случайных выборок из любого распределения (с конечной дисперсией) будет стремиться к нормальному распределению. Это справедливо вне зависимости от начального распределения данных.
- Закон больших чисел: по мере увеличения размера выборки среднее значение выборки приближается к математическому ожиданию генеральной совокупности. Это означает, что среднее значение выборки становится более точным представлением среднего значения всей популяции при увеличении количества наблюдений.
